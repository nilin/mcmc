```python
import pickle
data=dict()

for name in ['mcmc','correlated','no_mc']:
    with open(name,'rb') as f:
        data[name]=pickle.load(f)
```

```python
import matplotlib.pyplot as plt

def smooth(nums,r=.9):
    out=[]
    v_=nums[0]
    for v in nums:
        v_=v*(1-r)+v_*r
        out.append(v_)
    return out
    
for key,nums in data.items():
    data[key]=smooth(nums,0)

plt.plot(data['mcmc'],label='MCMC from both')
plt.plot(data['correlated'],label='correlated = MC from target')
plt.plot(data['no_mc'],label='No MC')

plt.semilogx()
plt.ylim(0,1)

plt.legend()
```
